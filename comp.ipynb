```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("elma_s2_s3.xlsx")
df.head()
```

```python
df.drop([0], inplace=True)
df.head()
```

```python
df_comp = df[df["game_type"] == "competitive"]
df_comp.reset_index(inplace=True, drop=True)
df_comp.head()
```

```python
all_columns = df_comp.columns
flt_columns = [col for col in all_columns if ("s2" not in col) and (("status" in col) or "p1" in col or "p2" in col or "game_id" in col or "s3_l1_player" in col or "game_name" in col)]

i = flt_columns.index("s3_l1_player")
to_move = flt_columns.pop(i)
flt_columns.insert(9, to_move)
flt_columns
```

```python
lvl_columns = [col for col in flt_columns if "status" in col]
lvl_columns
```

```python
df_flt = df_comp[flt_columns]
df_flt.head()
```

```python
game_IDs = np.unique(np.array(df_flt["game_id"].to_list()))
len(game_IDs)
```

```python
data = {}
for id in game_IDs:
    tmp = df_flt[df_flt["game_id"] == id]
    df_p1 = tmp[tmp["s3_l1_player"] == 1]
    df_p2 = tmp[tmp["s3_l1_player"] == 2]

    p1_id = np.unique(df_p1["p1_student_id"])[0]
    p2_id = np.unique(df_p2["p2_student_id"])[0]

    if p1_id == p2_id:
        continue

    p1_sch = np.unique(df_p1["school_p1"])[0]
    p2_sch = np.unique(df_p2["school_p2"])[0]
    p1_grade = np.unique(df_p1["p1_grade"])[0]
    p2_grade = np.unique(df_p2["p2_grade"])[0]
    p1_gender = np.unique(df_p1["p1_gender"])[0]
    p2_gender = np.unique(df_p2["p2_gender"])[0]
    game_name = np.unique(df_p1["game_name"])[0]

    p1_count = 0
    p2_count = 0
    for lvl in lvl_columns:
        if ("success" in df_p1[lvl].to_list()):
            p1_count += 1

        if ("success" in df_p2[lvl].to_list()):
            p2_count += 1
    
    if p1_count > p2_count:
        winner = "P1"
    elif p1_count == p2_count:
        winner = "Draw"
    else:
        winner = "P2"

    data[id] = {
        "P1": {"ID": p1_id, "Gender": p1_gender, "School": p1_sch, "Grade": p1_grade, "Game Name": game_name, "Maps Completed": p1_count},
        "P2": {"ID": p2_id, "Gender": p2_gender, "School": p2_sch, "Grade": p2_grade, "Game Name": game_name, "Maps Completed": p2_count},
        "Winner": winner
    }

data
```

```python
rows = []
for game_id, players, in data.items():
    row = {
        'game_id': game_id,
        'P1_ID': players['P1']['ID'],
        'P2_ID': players['P2']['ID'],
        'P1_gender': players['P1']['Gender'],
        'P2_gender': players['P2']['Gender'],
        'P1_school': players['P1']['School'],
        'P2_school': players['P2']['School'],
        'P1_grade': players['P1']['Grade'],
        'P2_grade': players['P2']['Grade'],
        'Game_name': players['P1']['Game Name'],
        'P1_maps_completed': players['P1']['Maps Completed'],
        'P2_maps_completed': players['P2']['Maps Completed'],
        'Winner': data[game_id]['Winner']
    }
    rows.append(row)

# Create DataFrame
df = pd.DataFrame(rows)

# Write to Excel
df.to_excel('game_data.xlsx', index=False)

print("Data has been written to game_data.xlsx")
```
